```python
import pandas as pd
```

```python
data1 = pd.read_excel("Lombard_Regular & TP_Brokerage_Apr'24(Statement).xlsx", sheet_name="RAW STATEMENT")
data2 = pd.read_excel("LOMBARD_REPORT_APRIL_24-25 TILL 07.06.2024(Saiba Dump).xls", engine='xlrd')
```

```python
# Assigning unique row names for data1 and data2
data1.index = ['L' + str(i+1) for i in range(len(data1))]
data2.index = ['S' + str(i+1) for i in range(len(data2))]
```

```python
# Specified columns to consider in data1 and data2
columns_data1 = ['CUSTOMER_ID', 'POL_NUM_TXT', 'INSURED_CUSTOMER_NAME', 'GWP', 'ACTUAL_COMMISSION', 'POLICY_END_DATE', 'POLICY_START_DATE']
columns_data2 = ['PolicyNo', 'CustName', 'OD Premium', 'PBST', 'Brokerage', 'TotalBrok.', 'Exp. Date', 'Policy_StartDate']
```

```python
# Initialize lists to store common row numbers
common_rows_data1 = []
common_rows_data2 = []

# Iterate through each specified column in data1
for col1 in columns_data1:
    if col1 in data1.columns:
        # Filter out NaN values and 0
        set1 = data1[col1].dropna().replace(0, None)
        
        # Iterate through each specified column in data2
        for col2 in columns_data2:
            if col2 in data2.columns:
                # Filter out NaN values and 0
                set2 = data2[col2].dropna().replace(0, None)
                
                # Common values and their row indices
                common_values = set(set1).intersection(set(set2))
                if common_values:
                    print(f"'{col1}' - '{col2}' [{len(common_values)} common values]")
                    for value in common_values:
                        row1_indices = data1[data1[col1] == value].index.tolist()
                        row2_indices = data2[data2[col2] == value].index.tolist()
                        for r1 in row1_indices:
                            for r2 in row2_indices:
                                common_rows_data1.append(r1)
                                common_rows_data2.append(r2)
                                print(f"{r1} -> {r2}")
```

```python
# Lists of common row numbers sorted in increasing order
print("\nCommon row numbers in data1:")
print(sorted(set(common_rows_data1), key=lambda x: int(x[1:])))

print("\nCommon row numbers in data2:")
print(sorted(set(common_rows_data2), key=lambda x: int(x[1:])))
```

```python
print(f"\nNumber of unique common row indices in data1: {len(set(common_rows_data1))}")
print(f"Number of unique common row indices in data2: {len(set(common_rows_data2))}")
```

```python
# Converting to sets for faster operations
common_rows_data1 = set(common_rows_data1)
common_rows_data2 = set(common_rows_data2)

# Identifying unmatched rows
unmatched_data1 = data1[~data1.index.isin(common_rows_data1)]
unmatched_data2 = data2[~data2.index.isin(common_rows_data2)]
```

```python
with pd.ExcelWriter("Unmatched_Data.xlsx") as writer:
    unmatched_data1.to_excel(writer, sheet_name='Unmatched_Data1')
    unmatched_data2.to_excel(writer, sheet_name='Unmatched_Data2')
```
